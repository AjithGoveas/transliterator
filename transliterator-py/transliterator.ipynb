```python
from indicnlp.tokenize.indic_tokenize import trivial_tokenize

text = "यह एक उदाहरण वाक्य है। यहाँ कुछ शब्द हैं।"
tokens = trivial_tokenize(text)

print(tokens)

detokenized_text = " ".join(tokens)
print(detokenized_text)
```

```python
# Sentence split
from indicnlp.tokenize.sentence_tokenize import sentence_split
sentences = sentence_split(text, lang='hi')
print(sentences)
```

```python
# Transliteration using indic-nlp-library
from indicnlp.transliterate.unicode_transliterate import UnicodeIndicTransliterator

# Transliterate from Devanagari to Latin
transliterated_text = UnicodeIndicTransliterator.transliterate(text, 'hi', 'bn')
print(transliterated_text)
```

```python
# Try aksharamukha too
from aksharamukha import transliterate
text = "नमस्ते विश्व"
aksharamukha_transliterated = transliterate.process('Devanagari', 'Tamil', text)
print(aksharamukha_transliterated)
```

```python
# Try with Location names
from aksharamukha import transliterate
import pandas as pd

location_names = [
	"Delhi",
	"Mumbai",
	"Kolkata",
	"Chennai",
	"Surathkal",
	"Police Station"
]

translated_names = []

for name in location_names:
	hindi_name = transliterate.process('ISO', 'Devanagari', name)
	bengali_name = transliterate.process('ISO', 'Bengali', name)
	kannada_name = transliterate.process('ISO', 'Kannada', name)
	tamil_name = transliterate.process('ISO', 'Tamil', name)
	translated_names.append((name, hindi_name, bengali_name, kannada_name, tamil_name))

df_locations = pd.DataFrame(translated_names, columns=['Original Name', 'Hindi (Devanagari)', 'Bengali', 'Kannada', 'Tamil'])
display(df_locations)
```
